# Multimodal Capabilities: Image Description

## Beyond Text - Vision-Language Models

Modern LLMs can process both text and images, enabling sophisticated multimodal applications.

### Vision Tasks Demonstrated:
- **Image Analysis**: Detailed description of visual content
- **Object Counting**: Quantitative analysis (number of people, genders)
- **Structured Output**: JSON-formatted responses for programmatic use

### Technical Implementation:
- **Base64 Encoding**: Convert images to text for API transmission
- **HumanMessage with Images**: LangChain's multimodal message format
- **JSON Mode**: Structured output for consistent parsing

### Use Cases:
- **Content Moderation**: Analyze image content automatically
- **Accessibility**: Generate descriptions for visually impaired users
- **Data Analysis**: Extract structured information from images

This section demonstrates practical image understanding with GPT-5.2.

````python
llm5 = ChatOpenAI(model="gpt-5.2")
````

````python
# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")
````

````python
img = encode_image(path)
````

````python
resp5 = llm5.invoke([
    HumanMessage(content=[
        {"type": "text", "text": "Qu'est ce que tu vois dans cette image"},
        {"type": "image_url", "image_url": {"url": f"data:image/png;base64,{img}"}}
    ])
])
````

````python
print(print(resp5.content))
````
